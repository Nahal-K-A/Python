```python
import math

def solve_quadratic(a, b, c):
    discriminant = b**2 - 4 * a * c
    if discriminant > 0:
        root1 = (-b + math.sqrt(discriminant)) / (2 * a)
        root2 = (-b - math.sqrt(discriminant)) / (2 * a)
        return f"Two real roots: {root1}, {root2}"

    elif discriminant <0:
        real_part = -b / (2 * a)
        imaginary_part = math.sqrt(-discriminant) / (2 * a)
        return f"Complex roots: {real_part}  {imaginary_part}i"
    else:
        root = -b / (2 * a)
        return f"One repeated root: {root}"
# Test values
for a, b, c in [(1, 5, 6), (1, 4, 4), (1, 2, 5)]:
    result = solve_quadratic(a, b, c)
    print(f"For a={a}, b={b}, c={c}: {result}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

V = 120  # Voltage source
Rs = 50  # Internal resistance of the source
Rl = np.linspace(1,500,100)
# Function to calculate power supplied to the load resistance
def power_supplied(Rl):
    return (V**2 * Rl) / ((Rs + Rl)**2)

  # Load resistance values from 1 to 100 o
power_values = power_supplied(Rl)

plt.plot(Rl, power_values)
plt.xlabel("Load Resistance (ohms)")
plt.ylabel("Power Supplied (watts)")
plt.title("Power Supplied to Load Resistance")
plt.grid(True)
plt.show()

# Maximum power occurs when Rl = Rs
max_power = max(power_values)
print(f"Maximum Power: {max_power} watts at Rl = Rs")
```

```python
import numpy as np
import matplotlib.pyplot as plt

I0 = 1e-9  # Reverse saturation current (1nA)
q = 1.602e-19  # Charge of an electron
k = 1.38e-23  # Boltzmann's constant
T = 300  # Temperature in Kelvin (e.g., 300K)

def diode_current(V):
    return I0 * (np.exp(q * V / (k * T)) - 1)

V_values = np.arange(0.3, 0.61, 0.01)
I_values = diode_current(V_values)

plt.plot(V_values, I_values)
plt.xlabel('Voltage (V)')
plt.ylabel('Current (A)')
plt.title('Diode Current vs Voltage')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Given constants
V = 120  # Voltage in Volts
R_s = 50  # Internal resistance in ohms

# Range of R_L values (load resistance)
R_L_values = np.linspace(0.1, 500, 500)  # Resistance from 0.1 ohm to 500 ohms

# List to store calculated power values
power_values = []

# For loop to calculate power for each R_L
for R_L in R_L_values:
    power = (V**2 / (R_s + R_L)**2) * R_L  # Power formula
    power_values.append(power)

# Plotting the results
plt.plot(R_L_values, power_values, label="Power vs Load Resistance")
plt.title("Power Supplied to Load Resistance")
plt.xlabel("Load Resistance (Ohms)")
plt.ylabel("Power (Watts)")
plt.grid(True)
plt.legend()
plt.show()
max_power = max(power_values)
print(f"Maximum Power: {max_power} watts at Rl = Rs")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
I0 = 1e-9  # Reverse saturation current (A)
q = 1.602e-19  # Electron charge (C)
k = 1.38e-23  # Boltzmann constant (J/K)

# Voltage range from 0.3V to 0.6V
voltages = np.arange(0.3, 0.61, 0.01)

# Temperatures in Kelvin
temperatures = [290, 310, 330]

# Plot for each temperature
plt.figure()

for T in temperatures:
    current = I0 * (np.exp(q * voltages / (k * T)) - 1)  # Diode current formula
    plt.plot(voltages, current, label=f"T = {T}K")

# Adding labels and title
plt.xlabel("Voltage (V)")
plt.ylabel("Current (A)")
plt.title("Diode Current vs Voltage for different temperatures")
plt.legend()
plt.grid(True)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Given values
R = 16e3  # resistance in ohms
C = 1e-6  # capacitance in farads

# Frequency range
frequencies = np.logspace(0, 6, num=1000)  # from 1 Hz to 1 MHz

# Low-pass filter response
G_lowpass = 1 / (1 + 1j * 2 * np.pi * frequencies * R * C)
amplitude_lowpass = np.abs(G_lowpass)
phase_lowpass = np.angle(G_lowpass)

# High-pass filter response
G_highpass = (1j * 2 * np.pi * frequencies * R * C) / (1 + 1j * 2 * np.pi * frequencies * R * C)
amplitude_highpass = np.abs(G_highpass)
phase_highpass = np.angle(G_highpass)

# Plotting
plt.figure(figsize=(12, 10))

# Low-pass filter amplitude and phase
plt.subplot(2, 2, 1)
plt.semilogx(frequencies, 20 * np.log10(amplitude_lowpass))
plt.title('Low-pass Filter Amplitude Response')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude (dB)')

plt.subplot(2, 2, 2)
plt.semilogx(frequencies, np.degrees(phase_lowpass))
plt.title('Low-pass Filter Phase Response')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Phase (degrees)')

# High-pass filter amplitude and phase
plt.subplot(2, 2, 3)
plt.semilogx(frequencies, 20 * np.log10(amplitude_highpass))
plt.title('High-pass Filter Amplitude Response')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude (dB)')

plt.subplot(2, 2, 4)
plt.semilogx(frequencies, np.degrees(phase_highpass))
plt.title('High-pass Filter Phase Response')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Phase (degrees)')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Function to compute the Fourier series for a square wave
def square_wave_fourier(t, n_terms):
    result = np.zeros_like(t)
    for n in range(n_terms):
        result += np.sin((2*n + 1) * np.pi * t) / (2*n + 1)
    return result * (4 / np.pi)

# Time range from -1.1 to 1.1 with steps of 0.01
t = np.arange(-1.1, 1.11, 0.01)

# Different numbers of terms to consider in the Fourier series
n_values = [1, 3, 5,7,9]

# Plotting
plt.figure(figsize=(10, 6))

for n in n_values:
    x_t = square_wave_fourier(t, n)
    plt.plot(t, x_t, label=f'n = {n}')

# Superimpose plots and add labels
plt.title('Fourier Series Approximation of a Square Wave')
plt.xlabel('t')
plt.ylabel('x(t)')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Given values
R = 16e3  # Resistance in ohms
C = 1e-6  # Capacitance in farads
f = np.logspace(1, 6, 1000)  # Frequency range from 10 Hz to 1 MHz
omega = 2 * np.pi * f  # Angular frequency

# Transfer functions
# Lowpass filter: G(f) = 1 / (1 + j*2*pi*f*R*C)
G_lowpass = 1 / (1 + 1j * omega * R * C)

# Highpass filter: G(f) = j*2*pi*f*R*C / (1 + j*2*pi*f*R*C)
G_highpass = (1j * omega * R * C) / (1 + 1j * omega * R * C)

# Amplitude response (magnitude)
amplitude_lowpass = np.abs(G_lowpass)
amplitude_highpass = np.abs(G_highpass)

# Phase response (angle in radians)
phase_lowpass = np.angle(G_lowpass)
phase_highpass = np.angle(G_highpass)

# Plotting the responses
plt.figure(figsize=(12, 8))

# Amplitude response plots
plt.subplot(2, 2, 1)
plt.semilogx(f, 20 * np.log10(amplitude_lowpass))
plt.title('Lowpass Filter Amplitude Response')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude (dB)')
plt.grid(True)

plt.subplot(2, 2, 2)
plt.semilogx(f, 20 * np.log10(amplitude_highpass))
plt.title('Highpass Filter Amplitude Response')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude (dB)')
plt.grid(True)

# Phase response plots
plt.subplot(2, 2, 3)
plt.semilogx(f, np.degrees(phase_lowpass))
plt.title('Lowpass Filter Phase Response')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Phase (degrees)')
plt.grid(True)

plt.subplot(2, 2, 4)
plt.semilogx(f, np.degrees(phase_highpass))
plt.title('Highpass Filter Phase Response')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Phase (degrees)')
plt.grid(True)

plt.tight_layout()
plt.show()
```
